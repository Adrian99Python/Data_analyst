We appropriately modify our data for smooth processing.

```python
df["Density\n(P/Km2)"] = df["Density\n(P/Km2)"].str.replace(',','').astype(float)
df['GDP'] = df['GDP'].str.replace('$','').str.replace(',','').astype(float)
df["Gasoline Price"] = df["Gasoline Price"].str.replace(',','').str.replace('$', '').astype(float)
df['Gross tertiary education enrollment (%)'] = df['Gross tertiary education enrollment (%)'].str.replace('%', '').astype(float)
df['CPI'] = df['CPI'].str.replace('.', '').str.replace(',', '').astype(float)
df['Minimum wage'] = df['Minimum wage'].str.replace('.', '').str.replace('$', '').astype(float)
df['Population'] = df['Population'].str.replace(',', '').astype(int)
df['Population: Labor force participation (%)'] = df['Population: Labor force participation (%)'].str.replace('%', '').astype(float)
df['Unemployment rate'] = df['Unemployment rate'].str.replace('%', '').astype(float)
```

```python
df.dtypes
```

We create the first correlation matrix primarily focused on the economic aspect.

```python
corelation = ["Density\n(P/Km2)","CPI","Gasoline Price","GDP","Gross tertiary education enrollment (%)","Minimum wage","Population: Labor force participation (%)","Unemployment rate","Population"]
corelation_matrix = df[corelation].corr()

plt.figure(figsize=(12,12))
sns.heatmap(corelation_matrix, annot=True, cmap="plasma", fmt=".2f", linewidth=0.5)
plt.title("Matrix of corelation")

plt.show()
```

```python
df["Agricultural Land( %)"] = df["Agricultural Land( %)"].str.replace('.','').str.replace('%','').astype(float)
df["Co2-Emissions"] = df["Co2-Emissions"].str.replace(',','').astype(float)
```

Next, we create another correlation matrix, this time focused on health.

```python
corelation_health = ["Density\n(P/Km2)","Population","Agricultural Land( %)","Birth Rate","Co2-Emissions","Fertility Rate","Infant mortality","Life expectancy","Maternal mortality ratio","Physicians per thousand"]
corelation_matrix_health = df[corelation_health].corr()

plt.figure(figsize=(12,12))
sns.heatmap(corelation_matrix_health, annot=True, cmap="magma", fmt=".2f", linewidth=0.5)
plt.title("Matrix of corelation of health")

plt.show()
```

In the next part, we investigate which country is responsible for the highest CO2 emissions.

```python
df_sorted = df.sort_values(by="Co2-Emissions", ascending=False)
df_top_20 = df_sorted.head(20)
df_top_20
```

We illustrate a chart depicting countries with the highest CO2 emissions.

```python
plt.figure(figsize=(14, 8))
plt.bar(df_top_20["Country"], df_top_20["Co2-Emissions"], color="tomato")
plt.xticks(rotation=50, ha='right')
plt.xlabel('Country')
plt.ylabel('Co2-Emissions')
plt.title('Top 20 Co2-Emissions Producers')
for index, value in enumerate(df_top_20["Co2-Emissions"]):
    label = str(value)[:2] 
    plt.text(index, value + 5, label, ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
df["Armed Forces size"] = df["Armed Forces size"].str.replace(',','').astype(float)
```

In the next section, we examine the percentage of population serving in the military for each country.

```python
Army = df[["Country", "Population", "Armed Forces size"]]
Army["Army in Population %"] = (df["Armed Forces size"] / df["Population"]) * 100
def add_percent_symbol(value):
    return f"{value:.2f}%"
Army["Army in Population %"] = Army["Army in Population %"].apply(add_percent_symbol)
Army
```

```python
Army["Army in Population %"] = Army["Army in Population %"].str.replace('%','').astype(float)
```

Here is the ranking of countries with the largest percentage of soldiers relative to their population.

```python
highest_armed_values = Army.nlargest(10, 'Army in Population %')
highest_armed_values['Army in Population %'] = highest_armed_values['Army in Population %'].astype(str)
highest_armed = highest_armed_values["Country"] + "  " + highest_armed_values['Army in Population %']
highest_armed
```

The following graph represents the sizes of armies in various countries around the globe.

```python
import plotly.express as px
fig = px.choropleth(df, locationmode="country names", locations=df['Country'], 
                    color="Armed Forces size",
                    title="Size of the Armed Forces For Each Country",
                    color_continuous_scale=px.colors.sequential.Inferno,
                    projection="natural earth")
fig.show()
```
