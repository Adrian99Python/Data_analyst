<center>Like we can see data in columns total_market_value and coach_name are not exist in our dataset.We can delete this columns from our frame.</center>

```python
clubs = clubs.drop(columns = ["total_market_value","coach_name"])
clubs.head()
```

<center>Next step is to sort our data by club_id , will be easier to work on .</center>

```python
clubs.sort_values(by= "club_id", inplace= True)
clubs
```

<center>Fill missing values in column foreigners_percentage. In our case we can replace column foreigners_percentage on resault of 
<b>(foreigners_number/squad_size)*100</b> .Additionaly we fillna values in average_age on mean values.</center>


```python
def calculate_foreigners_percentage(row):
    if row['foreigners_number'] == 0:
        return 0
    else:
        return round((row['foreigners_number'] / row['squad_size']) * 100, 2)

clubs["foreigners_percentage"] = clubs.apply(calculate_foreigners_percentage, axis=1)
clubs["average_age"].fillna(clubs['average_age'].mean(), inplace = True)
clubs
```

```python
# Results after fillna
clubs.isnull().sum()
```

<center>Let's find out which leagues are included in our DataFrame.</center>

```python
Leagues = clubs["domestic_competition_id"].unique()
Leagues
```

<center>Maybe adding new column with name of the Leagues will clarify data</center>

```python
### Creating a new Def for adding new column to DataFrame 
def liga(value):
    if value == "L1":
        return "Bundesliga"
    if value == "IT1":
        return "Serie A"
    if value == "TR1":
        return "Süper Lig"
    if value == "GB1":
        return "Premier League"
    if value == "ES1":
        return "La Liga"
    if value == "FR1":
        return "Ligue1"
    if value == "SC1":
        return "Scottish Premiership"
    if value == "GR1":
        return "Super League Greece"
    if value == "BE1":
        return "Belgian First Division A "
    if value == "RU1":
        return "Russian Premier League"
    if value == "NL1":
        return "Eredivisie"
    if value == "DK1":
        return "Danish Superliga"
    if value == "PO1":
        return "Primeira Liga"
    if value == "UKR1":
        return "Ukrainian Premier League"
    else:
        return "Diffrent League" 
```

```python
clubs["Leagues"] = clubs["domestic_competition_id"].apply(liga)
clubs = clubs[clubs.columns[-1:].tolist() + clubs.columns[:-1].tolist()]
clubs.head()
```

<h3><center><b>Data Analysis</b></center></h3>

Mean squad size for each league


```python
Mean_size_squad = clubs[["Leagues","squad_size"]]
Mean_size_squad
```

```python
average_size = round(Mean_size_squad.groupby('Leagues')['squad_size'].mean(),2)
average_size = average_size.sort_values(ascending=False)
average_size ### due to this information we can get information that bundesliga has the most numerous squad
```

Mean average age for each league

```python
Mean_average_age = clubs[["Leagues","average_age"]]
Mean_average_age
```

```python
average_age = round(Mean_average_age.groupby('Leagues')['average_age'].mean(),2)
average_age = average_age.sort_values(ascending=False)
average_age ### due to this information we can get information that la liga has eldest squads
```
